# 4.2 Functions ``*args`` and ``*kwargs``

```python
def add_all(*args):
    return sum(args)
add_all(1, 2, 3, 4)
```

```python
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(key, value)

print_info(name="Alice", age=30)
```

# 9 - Useful Python Features

```python
def log_call(func):
    def wrapper(*args, **kwargs):
        print("Function called")
        return func(*args, **kwargs)
    return wrapper


@log_call
def add(a, b):
    return a + b

print(add(3, 4))  
```

# Manipulation example

### Create a palindrome function

```python
def isPalindrome(word) : 
    word = word.lower().replace(" ", "")

    return word == word [::-1]

Palendrome = "radar"

isPalindrome(Palendrome)
```

### Create a function to check if the product of values in a table are 60

```python
def value_check(t): 
    for i in t:
        for j in t:
            for k in t:
                if i * j * k == 60:
                    return True
                
    return False

table = [2, 3, 4, 5]

value_check(table)
```

### Maximum value without max function

```python
def max(T): 
    maximum = T[0]
    for value in T:
        if value > maximum:
            maximum = value
    return maximum
```
